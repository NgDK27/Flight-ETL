```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from cassandrautils import *
```

```python
weather_df = getWeatherDF()
faker_df = getFakerDF()
flight_df = getFlightDF()

faker_flight_merged_df = pd.merge(faker_df, flight_df, how='left', left_on=['flight_origin', 'flight_destination'], right_on=['from_city', 'to_city'])
```

```python
faker_flight_merged_df
```

```python

# Create a bar chart of city of origin vs flight price
faker_flight_merged_df.groupby('flight_origin')['price_units'].mean().plot(kind='bar')
plt.xlabel('City of Origin')
plt.ylabel('Average Flight Price')
plt.title('City of Origin vs Average Flight Price')
plt.show()
```

```python
plt.figure(figsize=(20, 16))
sns.countplot(x='job', hue='is_cheapest', data=faker_flight_merged_df)
plt.xlabel('Job')
plt.ylabel('Count')
plt.title('Distribution of Cheapest Flights by Job')
plt.show()

#This chart will show the distribution of the is_cheapest field across different jobs. You can compare the heights of the bars to see if there's any noticeable difference between categories. 
# If there is, it might suggest a correlation between the categorical variable and whether or not a flight is the cheapest.
```

```python
weather_faker_merged_df = pd.merge(weather_df, faker_df, how='left', left_on='location', right_on='flight_origin')
```

```python
weather_faker_merged_df 
```

```python
plt.figure(figsize=(20, 16))
sns.boxplot(x='job', y='humidity', data=weather_faker_merged_df)
plt.xlabel('Job')
plt.ylabel('Humidity')
plt.title('Humidity Distribution at location by Job')
plt.show()

#This chart provides a narrative about the humidity at the locations for different job categories. 
# For example, if certain jobs have higher average temperatures or humidity, it might suggest that people in these professions are more likely to fly from warmer or more humid locations.
```
